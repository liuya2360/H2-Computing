```python
class Node(): 
    def __init__(self): 
        self.data = None 
        self.leftPtr = None 
        self.rightPtr = None 
    
    def setData(self, data): 
        self.data = data 
        
    def setLeftPtr(self, x): 
        self.leftPtr = x 
    
    def setRightPtr(self, y): 
        self.rightPtr = y 
    
    def getData(self): 
        return self.data 
    
    def getLeftPtr(self): 
        return self.leftPtr 
    
    def getRightPtr(self): 
        return self.rightPtr 
    
    def print(self): 
        print("Data:{}; Left:{}; Right:{}.".format(self.data, self.leftPtr, self.rightPtr))


class Tree(): 
    def __init__(self): 
        self.tree = []
        self.root = -1 
    
    def compare(self, a, b): 
        minLen = min(len(a), len(b))
        i = 0
        while minLen > 0: 
            minLen -= 1
            if ord(a[i]) < ord(b[i]): 
                return 1 #a should be in front of b 
            elif ord(a[i]) > ord(b[i]): 
                return -1 #b should be in front of a 
            else: 
                i += 1
            return 0 #a b are exactly the same 
        
    def add(self, newItem): 
        if self.root == -1: 
            self.root = 0
            newNode = Node()
            newNode.setData(newItem) 
            self.tree.append(newNode)
        else:  
            current = self.root
            while True: 
                if self.compare(self.tree[current].getData(), newItem) == -1: 
                    if self.tree[current].getLeftPtr() == None: 
                        newNode = Node() 
                        newNode.setData(newItem)
                        self.tree.append(newNode) 
                        self.tree[current].setLeftPtr(len(self.tree)-1) 
                        break 
                    else: 
                        current = self.tree[current].getLeftPtr() 
                else: 
                    if self.tree[current].getRightPtr() == None: 
                        newNode = Node() 
                        newNode.setData(newItem) 
                        self.tree.append(newNode) 
                        self.tree[current].setRightPtr(len(self.tree)-1)
                        break 
                    else: 
                        current = self.tree[current].getRightPtr() 
    
    def inOrderTraversalRec(self, current): 
        if current != None: 
            self.inOrderTraversalRec(self.tree[current].getLeftPtr()) 
            self.tree[current].print() 
            self.inOrderTraversalRec(self.tree[current].getRightPtr())
    
    def inOrderTraversal(self): 
        self.inOrderTraversalRec(self.root)
    
    def print(self): 
        for i in range(len(self.tree)): 
            self.tree[i].print()
        return 
    
newTree = Tree()
newTree.add("Dave")
newTree.add("Fred")
newTree.add("Ed") 
newTree.add("Greg")
newTree.add("Bob")
newTree.print()
print("\nInOrderTraversal Result: ")
newTree.inOrderTraversal()
```
