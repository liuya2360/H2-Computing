```python
#Task 1.1 
with open("DRINKS.TXT") as f: 
    line = f.read() 

data = line.strip().split("\n") 

class Drinklist(): 
    def __init__(self): 
        self.coffee = [] 
        self.tea = [] 
        self.other = []
    
    def read(self, data): 
        for row in data:
            if " " in row: 
                name = row.strip().split(" ") 
                temp = name[0]
            else: 
                temp = row 
                
            if temp.upper() == "KOPI": 
                self.coffee.append(row) 
            elif temp.upper() == "TEH": 
                self.tea.append(row) 
            else: 
                self.other.append(row) 
    
    def show_menu(self): 
        print("Menu:") 
        print("1. Brewed coffee") 
        print("2. Brewed tea") 
        print("3. Other drinks")
        flag = True 
        while flag: 
            select = input("Please select one type of drink: ") 
            if select == "1": 
                print("Brewed coffee: \n") 
                target = self.coffee
                flag = False 
            elif select == "2": 
                print("Brewed tea: \n") 
                target = self.tea 
                flag = False 
            elif select == "3": 
                print("Other drinks: \n") 
                target = self.other 
                flag = False 
            if flag: 
                continue 
            for i in target: 
                print(i) 
            print() 
            print("Total items: {}".format(len(target))) 
            
DL = Drinklist() 
DL.read(data) 
DL.show_menu() 
```

```python
#Task 2.1 

class hashTable(): 
    def __init__(self, size = 12): 
        self.array = [None for i in range(size)] 
        self.size = size 
    
    def hash(self, data): 
        return int(data) % 12 
    
    def insert(self, data): 
        self.array[self.hash(data)] = data  
        
    def print(self): 
        for row in self.array: 
            if row != None: 
                print(row) 
    
with open("KEYS.TXT") as f: 
    line = f.read() 
    
students = line.strip().split("\n") 

s = hashTable() 
for row in students: 
    s.insert(row) 
s.print() 
```

```python
#Task 2.2 

class hashTable2(): 
    def __init__(self, size = 12): 
        self.size = size 
        self.array = [None for i in range(size)] 
    
    def hash(self, data): 
        return int(data)%12 
    
    def insert(self, data): 
        target = self.hash(data) 
        while self.array[target] != None: 
            target = (target+1) % 12 
        self.array[target] = data 
    
    def print(self): 
        for row in self.array: 
            if row != None: 
                print(row) 
                
    def findAddress(self, data): 
        target = self.hash(data) 
        flag = False 
        while self.array[target] != data: 
            target = (target+1) % 12 
            flag = True 
            if flag and target == self.hash(data): 
                return "Student number not found. " 
        return target 

with open("KEYS2.TXT") as f: 
    line = f.read() 

students2 = line.strip().split("\n") 
                
s2 = hashTable2() 
for row in students2: 
    s2.insert(row) 
s2.print() 

#Task 2.3 
while True: 
    user_input = input("Please input the student number (enter E to exit):") 
    if user_input.upper() == 'E': 
        break 
    print(s2.findAddress(user_input)) 
```

```python
#Task 3.1  
class Node(): 
    def __init__(self, data): 
        self.data = data 
        self.leftPtr = None 
        self.rightPtr = None 
        
    def setData(self, s): 
        self.data = s 
    
    def setLeftPtr(self, x): 
        self.leftPtr = x 
    
    def setRightPtr(self, y): 
        self.rightPtr = y 
        
    def getData(self): 
        return self.data 
    
    def getLeftPtr(self): 
        return self.leftPtr 
    
    def getRightPtr(self): 
        return self.rightPtr 
        
class Tree(): 
    def __init__(self): 
        self.root = None 
        self.thisTree = [] 
    
    def add(self, newItem): 
        newNode = Node(newItem) 
        if self.root == None: 
            self.root = 0 
            self.thisTree.append(newNode)
        else: 
            current = self.root 
            while True: 
                if self.thisTree[current].getData() > newNode.getData(): 
                    if self.thisTree[current].getLeftPtr() == None: 
                        self.thisTree[current].setLeftPtr(len(self.thisTree)) 
                        self.thisTree.append(newNode) 
                        break 
                    else: 
                        current = self.thisTree[current].getLeftPtr() 
                else: 
                    if self.thisTree[current].getRightPtr() == None: 
                        self.thisTree[current].setRightPtr(len(self.thisTree)) 
                        self.thisTree.append(newNode) 
                        break 
                    else: 
                        current = self.thisTree[current].getRightPtr() 
                        
    def print(self): 
        for i in range(len(self.thisTree)): 
            print("Index: {}, data: {}, leftPtr: {}, rightPtr: {}".format(i, self.thisTree[i].getData(), self.thisTree[i].getLeftPtr(), self.thisTree[i].getRightPtr()))
            
    def postOrderTraversalRec(self, current): 
        if current != None: 
            self.postOrderTraversalRec(self.thisTree[current].getLeftPtr()) 
            self.postOrderTraversalRec(self.thisTree[current].getRightPtr()) 
            print("Index: {}, data: {}, leftPtr: {}, rightPtr: {}".format(current, self.thisTree[current].getData(), self.thisTree[current].getLeftPtr(), self.thisTree[current].getRightPtr()))
            
    def postOrderTraversal(self): 
        self.postOrderTraversalRec(self.root)
    
    
    
#Task 3.2 

t = Tree() 
t.add("Tiger") 
t.add("Lemur") 
t.add("Bat") 
t.add("Yak") 
t.add("Ostrich") 
t.add("Racc0on") 
t.add("Macaw") 
t.add("Zebra")
t.print() 

#Task 3.3 

print("Post order traversal: ") 
t.postOrderTraversal() 
```

```python
#Task 4.1 

def Pins(throw): 
    if throw == "X":
        return 10 
    else: 
        return int(throw) 
    
def Bowling_Score(throws): 
    return Bowling_Score_Helper(throws, 1) 

def Bowling_Score_Helper(throws, Frame_Num): 
    #frame 10 with no bonus 
    if Frame_Num == 10 and len(throws) == 2: 
        return Pins(throws[0]) + Pins(throws[1]) 
    #frame 10 with bonus 
    if Frame_Num == 10 and len(throws) == 3: 
        return Pins(throws[0]) + Pins(throws[1]) + Pins(throws[2])
    #strike 
    if throws[0] == "X": 
        Frame_Score = 10 + Pins(throws[1]) + Pins(throws[2])
        return Frame_Score + Bowling_Score_Helper(throws[1:], Frame_Num+1)
    
    Frame_Score = Pins(throws[0]) + Pins(throws[1]) 
    
    #a spare 
    if Frame_Score == 10: 
        return 10 + Pins(throws[2]) + Bowling_Score_Helper(throws[2:], Frame_Num+1) 
    
    return Frame_Score + Bowling_Score_Helper(throws[2:], Frame_Num+1)

print(Bowling_Score('X2815X91X365452X0X')) 
print(Bowling_Score('91739182X90X90X82X')) 

#Task 4.2 
```

```python
#Task 4.3 
with open("SCORES.TXT") as f: 
    line = f.read() 
    
raw = line.strip().split("\n") 
results = [] 
for row in raw: 
    row = row.strip().split(" ") 
    temp = [] 
    temp.append(row[0]) #register num 
    temp.append(row[1]) #country 
    score = 0
    for i in range(2, 8): 
        score += Bowling_Score(row[i]) 
    temp.append(score) 
    results.append(temp[:])

#bubble sort 
for i in range(len(results)-1): 
    swap = False 
    for j in range(len(results)-1-i): 
        if results[j][2] > results[j+1][2]: 
            swap = True 
            results[j][2], results[j+1][2] = results[j+1][2], results[j][2] 
    if not swap: 
        break 
            
#print 
print("Official Results") 
print("Position    Register Number    Country    Total Score") 
for i in range(len(results)): 
    print("    {:14}{:15}{:12}{}".format(str(i+1), results[i][0], results[i][1], results[i][2])) 
    
```
