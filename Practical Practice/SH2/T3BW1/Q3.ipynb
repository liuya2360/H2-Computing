```python
#Task 3.1 
class Node(): 
    def __init__(self, data): 
        self.data = data 
        self.next = None 

class DataStructure(): 
    def __init__(self):  
        self.head = None 
        self.tail = None 
    
    def isEmpty(self): 
        return self.head == None 
    
    def display(self): 
        if self.isEmpty(): 
            print("Linked list is empty. ") 
        else: 
            current = self.head 
            while current != None: 
                print(current.data) 
                current = current.next 
                
```

```python
#Task 3.2 

class Stack(DataStructure):
    def insert(self, value): 
        newNode = Node(value) 
        if self.isEmpty(): 
            self.head = newNode 
            self.tail = newNode 
        else: 
            self.tail.next = newNode 
            self.tail = newNode 
    
    def delete(self): 
        if self.isEmpty(): 
            return None 
        else: 
            toReturn = self.tail.data 
            if self.head == self.tail: 
                self.head = None 
                self.tail = None 
                return toReturn 
            else: 
                current = self.head 
                while current.next != self.tail: 
                    current = current.next 
                self.tail = current 
                self.tail.next = None 
                return toReturn 
            
    def displayRec(self, current): 
        if current == None: 
            return 
        self.displayRec(current.next) 
        print(current.data)
        
    def display(self): 
        self.displayRec(self.head) 
        
```

```python
#Task 3.3 

class Queue(DataStructure): 
    def insert(self, value): 
        newNode = Node(value) 
        if self.isEmpty(): 
            self.head = newNode 
            self.tail = newNode 
        else: 
            self.tail.next = newNode 
            self.tail = newNode 
    
    def delete(self): 
        if self.isEmpty(): 
            return None 
        else: 
            toReturn = self.head.data 
            if self.head == self.tail: 
                self.tail = None 
            self.head = self.head.next 
            return toReturn 
    
    def display(self): 
        current = self.head 
        while current.next != None: 
            print(current.data) 
            current = current.next 
            
```

```python
#Task 3.4 

with open("TASK3stack.txt") as f: 
    lineS = f.read() 
lineS = lineS.strip().split("\n") 
s = Stack() 
for i in lineS: 
    s.insert(i) 

with open("TASK3queue.txt") as f: 
    lineQ = f.read() 
lineQ = lineQ.strip().split("\n") 
q = Queue() 
for i in lineQ: 
    q.insert(i) 
    
print("Stack: ")
s.display() 
print("Queue: ") 
q.display() 
print("Remove two items from stack")
print(s.delete()) 
print(s.delete()) 
print("Remove two items from queue") 
print(q.delete()) 
print(q.delete()) 
print("Stack: ") 
s.display() 
print("Queue: ") 
q.display() 
```
